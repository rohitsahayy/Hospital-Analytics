```python
# Dropping Redundant Columns first. As can be seen above . only detail_df have 
# redundant/useless columns
cols_to_drop = ['Bed', 'Age Bucket', 'Custom', 'FZ me']
detail_df = detail_df.drop(columns=cols_to_drop)
```

```python
detail_df.columns
```

```python
# Applying some more Pre-processing tasks :

#Standardize Column Names

detail_df = detail_df.rename(columns={
    'Patient type':'Patient_Type',
    'treatemencost':'Treatment_Cost',
    'LOS':'Length_of_Stay'
})
```

```python
detail_df.columns
```

```python
#Gender Mapping
detail_df['Gender'] = detail_df['Gender'].str.upper().replace({
    'F':'Female',
    'M':'Male'
})
```

```python
#Clean Text Columns (Strip Spaces & Standardize Case)
for col in ['Name','Gender','City','State','Patient_Type','Status','Feedback']:
    detail_df[col] = detail_df[col].astype(str).str.strip().str.title()
```

.title() converts the text in a string to Title Case — meaning:

First letter of each word becomes uppercase

All other letters become lowercase

```python
#Convert Numeric Columns
detail_df['Age'] = pd.to_numeric(detail_df['Age'], errors='coerce')
detail_df['Treatment_Cost'] = pd.to_numeric(detail_df['Treatment_Cost'], errors='coerce')
detail_df['Length_of_Stay'] = pd.to_numeric(detail_df['Length_of_Stay'], errors='coerce')
detail_df['Rating'] = pd.to_numeric(detail_df['Rating'], errors='coerce')
```

errors='coerce'
Convert invalid values to NaN
safest for cleaning

```python
#Validate Foreign Key Links (important for SQLite)
detail_df = detail_df[detail_df['Staff_Id'].isin(staff_df['Staff_Id'])]
detail_df = detail_df[detail_df['Bed_ID'].isin(bed_df['Bed_ID'])]
detail_df = detail_df[detail_df['Dpt_ID'].isin(dept_df['Dpt_ID'])]
```

Creating the SQLite Database & Schema

```python
import sqlite3
```

```python
connection = sqlite3.connect("hospital_warehouse.db")
cursor = connection.cursor()
```

```python
cursor.executescript(
    """
    PRAGMA foreign_keys = ON;
    """
)
```

Tables i am going to create :
1) dim_patient : Information about patients
2) dim_staff : Information about staff/doctors
3) dim_department : Names of hospital departments
4) dim_bed : Bed descriptions/numbers/locations

5) fact_treatment : Stores each treatment event and its cost, duration, etc.

```python
dfs = {
    'detail_df': detail_df,
    'staff_df': staff_df,
    'dept_df': dept_df,
    'bed_df': bed_df
}

for name,df in dfs.items():
    print(f"Column in name {name}:")
    print(df.columns.tolist())
    print()
```

```python
command = """
DROP TABLE IF EXISTS fact_treatment;
DROP TABLE IF EXISTS dim_patient;
DROP TABLE IF EXISTS dim_staff;
DROP TABLE IF EXISTS dim_department;
DROP TABLE IF EXISTS dim_dept;
DROP TABLE IF EXISTS dim_bed;

CREATE TABLE dim_patient(
ID INTEGER PRIMARY KEY,
Name TEXT ,
Gender TEXT,
City TEXT,
State TEXT,
Age INTEGER,
Patient_Type TEXT
);

CREATE TABLE dim_staff(
Staff_Id INTEGER PRIMARY KEY,
Staff_Name TEXT
);

CREATE TABLE dim_dept(
Dpt_ID INTEGER PRIMARY KEY,
Department_Name TEXT
);

CREATE TABLE dim_bed(
Bed_ID INTEGER PRIMARY KEY,
Bed_Number TEXT
);

"""
```

```python
cursor.executescript(command)
```

```python
cursor.executescript(
    """
    DROP TABLE IF EXISTS fact_treatment;

    CREATE TABLE fact_treatment(
    Fact_ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Staff_Id INTEGER,
    Bed_ID INTEGER,
    Dpt_ID INTEGER,
    ID INTEGER,
    Status TEXT,
    Treatment_Cost REAL,
    Length_of_Stay REAL,
    ER_Time REAL,
    Date TEXT,
    Feedback TEXT,
    Rating REAL,

    FOREIGN KEY (Staff_Id) REFERENCES dim_staff(Staff_Id),
    FOREIGN KEY (ID) REFERENCES dim_patient(ID),
    FOREIGN KEY (Dpt_ID) REFERENCES dim_dept(Dpt_ID),
    FOREIGN KEY (Bed_ID) REFERENCES dim_bed(Bed_ID)
    );
    """
)
```
